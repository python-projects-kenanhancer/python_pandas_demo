```python
import pandas as pd

person_data: dict[str, list] = {
    "First_Name": ["John", "Alice", "Robert", "Emily", "Michael"],
    "Last_Name": ["Smith", "Johnson", "Williams", "Brown", "Davis"],
    "Age": [28, 34, 45, 23, 52],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],
}

person_df = pd.DataFrame(data=person_data)
```

```python
df_v1 = person_df[["First_Name", "Last_Name", "Age", "City"]]

df_v2 = person_df[["First_Name", "Last_Name", "Age", "City"]].copy()

df_v1, df_v2
```

```python
# Add new column to a DataFrame
def add_full_name(row: pd.Series) -> str:
    return row["First_Name"] + " " + row["Last_Name"]


person_df["Full_Name_v1"] = person_df.apply(add_full_name, axis=1)

person_df["Full_Name_v2"] = person_df.apply(lambda row: add_full_name(row), axis=1)

person_df["Full_Name_v3"] = person_df.apply(
    lambda row: row["First_Name"] + " " + row["Last_Name"], axis=1
)

person_df["Full_Name_v4"] = person_df["First_Name"] + " " + person_df["Last_Name"]

person_df = person_df.assign(
    Full_Name_v5=person_df["First_Name"] + " " + person_df["Last_Name"]
)

person_df
```

```python
df_v3 = person_df.assign(
    Full_Name_v6=person_df["First_Name"] + " " + person_df["Last_Name"]
)

full_name_series_v1: pd.Series = person_df.apply(
    lambda row: row["First_Name"] + " " + row["Last_Name"], axis=1
)

full_name_series_v2 = person_df["First_Name"] + " " + person_df["Last_Name"]

df_v3["Full_Name_v7"] = full_name_series_v1

df_v3["Full_Name_v8"] = full_name_series_v2

df_v3
```

```python
full_name_series_v3: pd.Series = person_df.apply(
    lambda row: row["First_Name"] + " " + row["Last_Name"], axis=1
)

df_v4 = person_df.assign(
    Full_Name_v9=full_name_series_v3,
    Full_Name_v10=person_df.apply(
        lambda row: row["First_Name"] + " " + row["Last_Name"], axis=1
    ),
    Full_Name_v11=lambda row: row["First_Name"] + " " + row["Last_Name"],
    Full_Name_v12=person_df["First_Name"] + " " + person_df["Last_Name"],
).assign(Full_Name_v13=lambda row: row["First_Name"] + " " + row["Last_Name"])

df_v4
```

```python
df_v5 = (
    person_df.copy()
    .assign(Full_Name_v6=lambda row: row["First_Name"] + " " + row["Last_Name"])
    .assign(Full_Name_v7=person_df["First_Name"] + " " + person_df["Last_Name"])
)

df_v5
```

```python
# DataFrame Creation:
# pandas: pd.DataFrame(data)
# PySpark: spark.createDataFrame(data)

# Selecting Columns:
# pandas: df['column_name']
# PySpark: df.select('column_name')

# Adding New Columns:
# pandas: df['new_column'] = df['existing_column'] * 2
# PySpark: df = df.withColumn('new_column', df['existing_column'] * 2)

# Map/Apply Operations:
# pandas (Map): df['column'].map(lambda x: func(x)) for element-wise operation on a Series.
# pandas (Apply): df.apply(lambda row: func(row), axis=1) for row-wise operation on a DataFrame.
# PySpark: Use withColumn along with a UDF for element-wise or row-wise transformations.
# Define a UDF: func_udf = F.udf(lambda x: func(x))
# Apply UDF: df = df.withColumn('new_column', func_udf(df['existing_column']))

# Filtering Rows:
# pandas: df[df['column'] > value]
# PySpark: df.filter(df['column'] > value)

# GroupBy Aggregations:
# pandas: df.groupby('column').agg({'other_column': 'sum'})
# PySpark: df.groupBy('column').agg(F.sum('other_column'))

# Joining DataFrames:
# pandas: df1.merge(df2, on='key_column')
# PySpark: df1.join(df2, df1['key_column'] == df2['key_column'])

# Sorting Data:
# pandas: df.sort_values(by='column')
# PySpark: df.orderBy('column')

# Handling Missing Values:
# pandas: df.fillna(value)
# PySpark: df.na.fill(value)

# Collecting Data:
# pandas: Data is already in local memory, so operations like to_list or direct iteration are possible.
# PySpark: df.collect() to bring the distributed data to the local machine (use with caution due to memory constraints).
```

```python
# Filtering Rows based on Age > 30
filtered_df = person_df[person_df["Age"] > 30]

filtered_df_v2 = person_df.filter(person_df["Age"] > 30)

filtered_df_v3 = person_df[person_df["Age"] > 30].copy()

filtered_df_v4 = person_df.filter(person_df["Age"] > 30).copy()

filtered_df, filtered_df_v2, filtered_df_v3, filtered_df_v4
```

```python
# Selecting Specific Columns
selected_columns_df = person_df[["First_Name", "Last_Name", "Age"]]

selected_columns_df_v2 = person_df[["First_Name", "Last_Name", "Age"]].copy()

selected_columns_df, selected_columns_df_v2
```

```python
# Calculating Statistics on Numeric Columns
age_min = person_df["Age"].min()
age_max = person_df["Age"].max()
age_mean = person_df["Age"].mean()
age_sum = person_df["Age"].sum()
age_median = person_df["Age"].median()
age_std = person_df["Age"].std()

age_min, age_max, age_mean, age_sum, age_median, age_std

# Generate tuple
age_stats_tuple = (age_min, age_max, age_mean, age_sum, age_median, age_std)

# Generate list
age_stats_list = [age_min, age_max, age_mean, age_sum, age_median, age_std]

# Generate dictionary
age_stats_dict = {
    "min": age_min,
    "max": age_max,
    "mean": age_mean,
    "sum": age_sum,
    "median": age_median,
    "std": age_std,
}

age_stats_tuple, age_stats_list, age_stats_dict
```
